```python
df_1=pd.read_csv("data_1.csv",sep="\t",names=["user_id","item_id","rating","timestamp"])
```

```python
df_2=pd.read_csv("data_2.csv")
```

```python
df_1
```

```python
df_2
```

```python
jovian.commit()
```

## merging  both the dataset

```python
df_1=pd.merge(df_1,df_2,on="item_id")
```

```python
df_1
```

```python
movie_df=df_1.pivot_table(index="user_id",columns="title",values="rating")
```

```python
movie_df
```

```python
ratings_df=pd.DataFrame(df_1.groupby("title")["rating"].mean())
```

```python
ratings_df
```

```python
ratings_df["number_of_ratings"]=pd.DataFrame(df_1.groupby("title")["rating"].count())
```

```python
ratings_df
```

```python
starwars_user_rating=movie_df["Star Wars (1977)"]
```

```python
starwars_user_rating
```

```python
corr_sw=movie_df.corrwith(starwars_user_rating)
```

```python
corr_sw
```

```python
corr_sw_df=pd.DataFrame(corr_sw,columns=["correlation"])
```

```python
corr_sw_df
```

```python
corr_sw_df.dropna(inplace=True)
```

```python
corr_sw_df.sort_values("correlation",ascending=False).head(10)
```

```python
corr_sw_df=corr_sw_df.join(ratings_df["number_of_ratings"])
```

```python
corr_sw_df
```

#### select those movies whose rating count is above 100.

```python
corr_sw_df[corr_sw_df["number_of_ratings"]>100].sort_values("correlation",ascending=False).head(10)
```
